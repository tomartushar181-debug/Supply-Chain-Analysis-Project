### Business Problem


##### A global e-commerce company operating across multiple regions manages end-to-end order fulfillment, including shipping and delivery, for products like sportin ggoods. The company is facing inconsistent delivery performance, where actual shipping times often deviate from scheduled timelines, leading to late deliveries and unpredictable order profitability.

### Desired Outcome

##### The goal is to analyze delivery operations, identify bottlenecks, and build a predictive system to reduce delays, optimize shipping decisions, and improve overallprofitability and efficiency.

```python
# Calculate order processing time (Actual shipping days)
df['Order Processing Time'] = (
    df['shipping date (DateOrders)'] - df['order date (DateOrders)']
).dt.days

# Calculate delay
df['Delay'] = (
    df['Order Processing Time'] - df['Days for shipment (scheduled)']
)

# Check if the order was delayed
df['Is_Delayed'] = df['Delay'] > 0

# Extract order month
df['order_month'] = df['order date (DateOrders)'].dt.month

# Extract day name
df['order_day'] = df['order date (DateOrders)'].dt.day_name()

# Extract order hour
df['order_hour'] = df['order date (DateOrders)'].dt.hour

# Summary statistics
df.describe()
```

```python
# Function to format large money values
def format_func(value):
    # Convert to Millions
    if value >= 1e6:
        return f'{value/1e6:.1f}M $'

    # Convert to Thousands
    elif value >= 1e3:
        return f'{value/1e3:.1f}K $'

    # Show normal value
    else:
        return f'{value:.0f} $'


# Keep only delayed orders
delayed_df = df[df['Delay'] > 0]

# Create dictionary to store KPIs
metrics = {}

# Total number of orders
metrics['Total Orders'] = len(df)

# Total delayed orders
metrics['Late Deliveries'] = len(delayed_df)

# 90th percentile of delay
metrics['90% Delay (days)'] = delayed_df['Delay'].quantile(0.90)

# On-time delivery percentage
metrics['On Time Delivery (%)'] = (
    1 - metrics['Late Deliveries'] / metrics['Total Orders']
) * 100

# Late delivery percentage
metrics['Late Delivery (%)'] = (
    metrics['Late Deliveries'] / metrics['Total Orders']
) * 100

# Total profit from profitable orders
metrics['Total Profit'] = format_func(
    df.loc[
        df['Order Profit Per Order'] > 0,
        'Order Profit Per Order'
    ].sum()
)

# Profit from delayed orders
# (Label can be changed if you actually mean profit, not loss)
metrics['Total Loss due to delays'] = format_func(
    df.loc[
        df['Delay'] > 0,
        'Order Profit Per Order'
    ].sum()
)

# Print KPIs
print("\n--- Business KPIs ---\n")

for k, v in metrics.items():
    if isinstance(v, float):
        print(f"{k}: {v:.2f}")
    else:
        print(f"{k}: {v}")
```

### Bottleneck detection
###### Bottleneck detection means finding the part of a process that slows everything else down.

```python
# Function
def delay_chart(category, ax):

    # Total Orders
    total_orders = (
        df.groupby(category)['Is_Delayed']
          .count()
          .reset_index(name='Total Orders')
    )

    # Late Orders
    late_orders = (
        df.groupby(category)['Is_Delayed']
          .sum()
          .reset_index(name='Late Orders')
    )

    # Merge both tables
    delay_df = pd.merge(total_orders, late_orders, on=category)

    # Calculate Delay %
    delay_df['Delay %'] = (
        delay_df['Late Orders'] /
        delay_df['Total Orders']
    ) * 100

    # Sort and keep Top 10
    delay_df = delay_df.sort_values(
        by='Delay %',
        ascending=False
    ).head(10)

    # Bar Chart
    sns.barplot(
        data=delay_df,
        x='Delay %',
        y=category,
        palette='viridis',
        ax=ax
    )

    # Add extra space for labels
    ax.set_xlim(0, delay_df['Delay %'].max() + 8)

    # Titles
    ax.set_title(f'Delay % by {category}', fontsize=12)
    ax.set_xlabel('Delay Percentage (%)')
    ax.set_ylabel(category)

    # Data Labels
    for i, value in enumerate(delay_df['Delay %']):

        # Large bars â label inside
        if value >= 12:
            ax.text(
                value - 1,
                i,
                f'{value:.1f}%',
                ha='right',
                va='center',
                color='white',
                fontsize=9,
                fontweight='bold'
            )

        # Small bars â label outside
        else:
            ax.text(
                value + 0.5,
                i,
                f'{value:.1f}%',
                ha='left',
                va='center',
                color='black',
                fontsize=9,
                fontweight='bold'
            )


# ===============================
# Create 2 Ã 3 Dashboard
# ===============================

fig, ax = plt.subplots(2, 3, figsize=(18, 10))

delay_chart('Order Region', ax[0,0])
delay_chart('Customer Segment', ax[0,1])
delay_chart('Shipping Mode', ax[0,2])

delay_chart('Order Status', ax[1,0])
delay_chart('Type', ax[1,1])
delay_chart('Department Name', ax[1,2])

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Step 1 : Keep only Central Africa orders
# -----------------------------
central_df = df[df['Order Region'] == 'Central Africa']

# -----------------------------
# Shipping Mode
# -----------------------------
shipping = (
    central_df.groupby('Shipping Mode')['Is_Delayed']
    .agg(
        Total_Orders='count',
        Late_Orders='sum'
    )
    .reset_index()
)

shipping['Delay %'] = shipping['Late_Orders'] / shipping['Total_Orders'] * 100
shipping['Driver'] = 'Shipping Mode'
shipping['Factor'] = 'Shipping Mode : ' + shipping['Shipping Mode']

shipping = shipping[['Driver','Factor','Delay %','Total_Orders']]

# -----------------------------
# Customer Segment
# -----------------------------
customer = (
    central_df.groupby('Customer Segment')['Is_Delayed']
    .agg(
        Total_Orders='count',
        Late_Orders='sum'
    )
    .reset_index()
)

customer['Delay %'] = customer['Late_Orders'] / customer['Total_Orders'] * 100
customer['Driver'] = 'Customer Segment'
customer['Factor'] = 'Customer Segment : ' + customer['Customer Segment']

customer = customer[['Driver','Factor','Delay %','Total_Orders']]

# -----------------------------
# Department Name
# -----------------------------
department = (
    central_df.groupby('Department Name')['Is_Delayed']
    .agg(
        Total_Orders='count',
        Late_Orders='sum'
    )
    .reset_index()
)

department['Delay %'] = department['Late_Orders'] / department['Total_Orders'] * 100
department['Driver'] = 'Department Name'
department['Factor'] = 'Department Name : ' + department['Department Name']

department = department[['Driver','Factor','Delay %','Total_Orders']]

# -----------------------------
# Order Status
# -----------------------------
status = (
    central_df.groupby('Order Status')['Is_Delayed']
    .agg(
        Total_Orders='count',
        Late_Orders='sum'
    )
    .reset_index()
)

status['Delay %'] = status['Late_Orders'] / status['Total_Orders'] * 100
status['Driver'] = 'Order Status'
status['Factor'] = 'Order Status : ' + status['Order Status']

status = status[['Driver','Factor','Delay %','Total_Orders']]

# -----------------------------
# Type
# -----------------------------
payment = (
    central_df.groupby('Type')['Is_Delayed']
    .agg(
        Total_Orders='count',
        Late_Orders='sum'
    )
    .reset_index()
)

payment['Delay %'] = payment['Late_Orders'] / payment['Total_Orders'] * 100
payment['Driver'] = 'Type'
payment['Factor'] = 'Type : ' + payment['Type']

payment = payment[['Driver','Factor','Delay %','Total_Orders']]

# -----------------------------
# Combine all tables
# -----------------------------
final_df = pd.concat(
    [shipping, customer, department, status, payment],
    ignore_index=True
)

# -----------------------------
# Optional:
# Remove categories with very few orders
# -----------------------------
final_df = final_df[final_df['Total_Orders'] >= 20]

# -----------------------------
# Top 10 Drivers
# -----------------------------
top10 = (
    final_df.sort_values(
        by='Delay %',
        ascending=False
    )
    .head(10)
)

print(top10)

# -----------------------------
# Plot
# -----------------------------
plt.figure(figsize=(10,6))

bars = plt.barh(
    top10['Factor'],
    top10['Delay %'],
    color='slateblue'
)

plt.title("Top Drivers of Late Delivery in Central Africa")
plt.xlabel("Delay Percentage (%)")
plt.ylabel("Driver Factors")

plt.gca().invert_yaxis()

for bar in bars:
    width = bar.get_width()
    plt.text(
        width - 3,
        bar.get_y() + bar.get_height()/2,
        f"{width:.1f}%",
        color='white',
        va='center'
    )

plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.show()
```

```python
# Delay % by Month, Day of Week, Hour

delay_by_month = (
df.groupby('order_month') ['Is_Delayed']
.mean()
.reset_index()
)
delay_by_month['delay_pct'] = delay_by_month['Is_Delayed'] * 100

delay_by_day = (
df.groupby('order_day') ['Is_Delayed']
.mean()
.reset_index()
)
delay_by_day['delay_pct'] = delay_by_day['Is_Delayed'] * 100

delay_by_hour = (
df.groupby('order_hour') ['Is_Delayed']
.mean()
.reset_index()
)
delay_by_hour['delay_pct'] = delay_by_hour['Is_Delayed'] * 100
```

```python
import matplotlib.pyplot as plt

# Create figure and subplots (3 side-by-side)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# =========================
# 1. Delay % by Month (Line)
# =========================
axes[0].plot(delay_by_month['order_month'],
             delay_by_month['delay_pct'],
             marker='o')

# Add data labels (Month chart)
for x, y in zip(delay_by_month['order_month'], delay_by_month['delay_pct']):
    axes[0].annotate(f'{y:.1f}%',
                     (x, y),
                     textcoords="offset points",
                     xytext=(0, 8),
                     ha='center')

axes[0].set_title('Delay % Trend Over Month')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Delay Percentage (%)')
axes[0].tick_params(axis='x', rotation=45)

# =========================
# 2. Delay % by Day (Bar)
# =========================
bars = axes[1].bar(delay_by_day['order_day'],
                    delay_by_day['delay_pct'])

# Add data labels (Day chart)
for bar in bars:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2,
                 height,
                 f'{height:.1f}%',
                 ha='center',
                 va='bottom')

axes[1].set_title('Delay % by Day of Week')
axes[1].set_xlabel('Day of Week')
axes[1].set_ylabel('Delay Percentage (%)')
axes[1].tick_params(axis='x', rotation=45)

# =========================
# 3. Delay % by Hour (Line)
# =========================
axes[2].plot(delay_by_hour['order_hour'],
             delay_by_hour['delay_pct'],
             marker='o')

# Add data labels (Hour chart)
for x, y in zip(delay_by_hour['order_hour'], delay_by_hour['delay_pct']):
    axes[2].annotate(f'{y:.1f}%',
                     (x, y),
                     textcoords="offset points",
                     xytext=(0, 8),
                     ha='center')

axes[2].set_title('Delay % by Hour')
axes[2].set_xlabel('Hour of Day')
axes[2].set_ylabel('Delay Percentage (%)')

# Layout fix
plt.tight_layout()

# Show plot
plt.show()
```
